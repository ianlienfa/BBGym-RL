```python
def read_loss_file(filename):  
    f = open(filename, "r")
    str = f.readline()
    li = list(str.split(", "))
    li = li[:-1]
    li = [float(i) for i in li]
    return li
```

```python
import numpy as np
from matplotlib import pyplot as plt
train_process_track = ["../build/kdypnmpsxl.in.bt", "../build/2.txt.bt", "../build/anfgjhdqwx.in.bt"]
    
def train_progress_tracker():
    size = 2
    fig, ax = plt.subplots(size, size, figsize=(20, 20))
    x, y, ct = 0, 0, 0
    for filename in train_process_track:        
        net_node_searched = [int(x) for x in (open(filename).read().splitlines())]
        print(net_node_searched)
        ax[y][x].plot(net_node_searched)
        ax[y][x].set_title(filename)
        ct += 1
        x += 1
        if x >= size:
            x, y = 0, y+1

train_progress_tracker()
```

```python
import os
from matplotlib import pyplot as plt
from operator import itemgetter

def read_node_searched_file(filename):  
    f = open(filename, "r")
    net_node_searched = int(f.readline())
    bfs_node_searched = int(f.readline())    
    level_node_searched = int(f.readline())    
    rand_node_searched = int(f.readline())    
    return net_node_searched, bfs_node_searched, level_node_searched, rand_node_searched

def testing(directory_name):
    size = 5
    fig, ax = plt.subplots(size, size, figsize=(20, 20))
    x, y, ct = 0, 0, 0
    for filename in os.listdir(directory_name):
        net_node_searched, bfs_node_searched, level_node_searched, rand_node_searched = read_node_searched_file(directory_name + filename)                
        keys = ['net', 'bfs', 'level', 'rand']
        color = ['cyan', 'cyan', 'cyan', 'cyan']        
        values = [net_node_searched, bfs_node_searched, level_node_searched, rand_node_searched]
        index, element = min(enumerate(values), key=itemgetter(1))        
        color[index] = 'red'
        ax[y][x].bar(keys, values, color=color)
        ax[y][x].set_title(filename)
        ct += 1
        x += 1
        if x >= size:
            x, y = 0, y+1
```
