```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def read_loss_file(filename):  
    f = open(filename, "r")
    str = f.readline()
    li = list(str.split(", "))
    li = li[:-1]
    li = [float(i) for i in li]
    return li
```

```python
pi_loss = read_loss_file("../saved_model/pi_loss.txt")
q_loss = read_loss_file("../saved_model/q_loss.txt")
```

```python
# plot two figures
plt.plot(pi_loss)
plt.ylabel('pi loss')
plt.show()
```

```python
# plot two figures
plt.plot(q_loss)
plt.ylabel('q loss')
plt.show()
```
