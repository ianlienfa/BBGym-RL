```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def read_loss_file(filename):  
    f = open(filename, "r")
    str = f.readline()
    li = list(str.split(", "))
    li = li[:-1]
    li = [float(i) for i in li]
    return li
```

```python
pi_loss = read_loss_file("../saved_model/pi_loss.txt")
q_loss = read_loss_file("../saved_model/q_loss.txt")
rewards = read_loss_file("../saved_model/rewards.txt")
```

```python
# plot two figures
plt.plot(pi_loss)
plt.ylabel('pi loss')
plt.show()
```

```python
raw_rewards = rewards
rewards = []
i = 10
tmp = 0
for it in range(len(raw_rewards)):
    if(i != 0):
        tmp += raw_rewards[it]
    else:
        rewards.append(tmp)
        tmp = 0
        i = 10
    i -= 1

plt.plot(rewards)
axis = plt.gca()
axis.set_ylim([-3, 3])
plt.ylabel('rewards')
plt.show()
```
